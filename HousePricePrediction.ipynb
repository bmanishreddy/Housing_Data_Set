```python
# to handle datasets
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt
% matplotlib inline

# to display all the columns of the dataframe in the notebook
pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('train.csv')
```

```python
print(data.shape)
```

```python
data.head()
```

```python
with_vars_na = [var for var in data.columns if data[var].isnull().sum()>1]
```

```python
print(with_vars_na)
```

```python
#Check the null values 
```

```python
for var in with_vars_na:
    print(var, np.round(data[var].isnull().mean(),3),'%missing values ')
```

```python
def AnalyseData(df, var):
    df = df.copy()
    
    df[var] = np.where(df[var].isnull(), 1, 0)
    
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.show()
    
for var in with_vars_na:
    AnalyseData(data, var)
```

```python
#Get the set of numerical varibles 
```

```python
num_vars = [var for var in data.columns if data[var].dtypes != 'O']
print("The number of Numericals = ",len(num_vars))

data[num_vars].head()
```

```python
print("No of house id labels  = ",len(data.Id.unique()))
print("No of houses in data set =",len(data))
```

```python
year_vars = [var for var in num_vars if 'Yr' in var or 'Year' in var]

year_vars
```

```python
for var in year_vars:
    print(var,data[var].unique())
    print()
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
```

```python
def Year_sale(df, var):
    df = df.copy()
    df[var] = df['YrSold'] - df[var]
    plt.scatter(df[var],df['SalePrice'])
    plt.ylabel('SalePrice')
    plt.xlabel(var)
    plt.show()
for var in year_vars:
    if var != 'YrSold':
        Year_sale(data,var)
```

```python
descreteVaribles = [var for var in num_vars if len(data[var].unique())<20 and var not in year_vars+['Id']]
print('The number of descrete varibles = ',len(descreteVaribles))
```

```python
data[descreteVaribles].head()
```

```python
def AnalyseDescreteVaribles(df,var):
    df = df.copy()
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylabel('SalePrice')
    plt.show()
    
for var in descreteVaribles:
    AnalyseDescreteVaribles(data,var)
```

```python
#continious varibles 
```

```python
cont_vars= [var for var in num_vars if var not in descreteVaribles+year_vars+['Id']]
print("no of vars =",len(cont_vars))
```

```python
data[cont_vars].head()
```

```python
def analyse_cont(df,var):
    df = df.copy()
    df[var].hist(bins = 20)
    plt.ylabel("Number of houses ")
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in cont_vars:
    analyse_cont(data,var)
```
